```python
# Import packages for use in notebook

import os
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy.interpolate import griddata
from scipy.interpolate import Rbf

%matplotlib inline
```

```python
os.chdir("/Users/euan-soutter/Desktop/Python/Field_Data/Azerbaijan/Data") # Sets working directory
df = pd.read_csv("az_beds_mm_avg.csv") # Loads data from working directory and assigns name 'df'
df.dropna()
df.describe() # Displays statistical descriptions of df
```

```python
extent = df.long.min(), df.long.max(), df.lat.min(), df.lat.max() # Calls X and Y from loaded data for grid extent
extent = [48.245278, 48.268674, 41.195161, 41.20455] # Sets x and y limits of grid (rounded)
grid_long, grid_lat = np.mgrid[48.245278:48.268674:0.000001, 41.195161:41.20455:0.000001] # Makes mesh grid of X and Y and assigns grid to grid_X and grid_Y                     

rbf = Rbf(df.long, df.lat, df.z)
rbf1 = Rbf(df.long, df.lat, df.meanthickness)
rbf2 = Rbf(df.long, df.lat, df.meangsize)

fig, axs = plt.subplots(ncols=3, figsize=(11, 5)) # Plots figure and subplots within figure
cmap = 'viridis' # Sets color bar
function = 'linear' # Sets interpolation method

ax = axs[0] # Assigns axis to position within figure
rbf = Rbf(df.long, df.lat, df.z, function=function) # Interpolates scattered data 
di = rbf(grid_long, grid_lat) # Assingns interpolated values to 'di'
cax1 = ax.imshow(di.T, origin="lower", extent=extent, cmap=cmap, aspect='auto', vmin=1000, vmax=4000) # Plots interpolation
ax.scatter(df.long, df.lat, s=5, c='w') # Plots scatter of sample points
ax.set_title('Height', pad=15) # Sets title of subplot
ax.set_ylabel('Latitude') # Sets y label of subplot
ax.set_xlabel('Longitude') # Sets x label of subplot
# ax.set_ylim(bottom=df.lat.min(),top=df.lat.max())
# ax.set_xlim(left=df.long.min(),right=df.long.max())
ax.ticklabel_format(useOffset=False, style='plain')
cbar = fig.colorbar(cax1, ax=axs[0], orientation='vertical') # Plots color bar
cbar.ax.set_ylabel('Height (m)', rotation=270, labelpad=15) # Plots labels and design of colorbar
# ax.tricontour(df.X, df.Y, df.mud, linewidths=0.5, colors='k') # Plots contours

ax = axs[1]
rbf1 = Rbf(df.long, df.lat, df.meanthickness, function=function)
di = rbf1(grid_long, grid_lat)
cax2 = ax.imshow(di.T, origin="lower", extent=extent, cmap=cmap, aspect='auto',vmin=0, vmax=100)
ax.scatter(df.long, df.lat, s=5, c='w')
ax.set_title('Mean Thickness', pad=15)
ax.set_ylabel('Latitude')
ax.set_xlabel('Longitude')
ax.ticklabel_format(useOffset=False, style='plain')
cbar = fig.colorbar(cax2, ax=axs[1], orientation='vertical')
cbar.ax.set_ylabel('Mean Thickness (m)', rotation=270,labelpad = 15)
# ax.tricontour(df.X, df.Y, df.silt, linewidths=0.5, colors='k')

ax = axs[2]
rbf2 = Rbf(df.long, df.lat, df.meangsize, function=function)
di = rbf2(grid_long, grid_lat)
cax3 = ax.imshow(di.T, origin="lower", extent=extent, cmap=cmap, aspect='auto', vmin=2, vmax=7)
ax.scatter(df.long, df.lat, s=5, c='w')
ax.set_title('Mean Grain Size', pad=15)
ax.set_ylabel('Latitude')
ax.set_xlabel('Longitude')
ax.ticklabel_format(useOffset=False, style='plain')
cbar = fig.colorbar(cax3, ax=axs[2], orientation='vertical')
cbar.ax.set_ylabel('Mean Grain Size', rotation=270, labelpad=15)
# ax.tricontour(df.X, df.Y, df.vfsand, linewidths=0.5, colors='k')

fig.suptitle('Measured Cenomanian-Turonian Beds', y=1.05,fontsize=14)
plt.tight_layout(w_pad=2)
plt.savefig('beds_minus_mud.png', dpi=400, bbox_inches='tight')  # Saves figure to working directory
```
