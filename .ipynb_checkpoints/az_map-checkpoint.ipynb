```python
import pandas as pd
from ipyleaflet import Map, Marker, Popup, basemaps, ImageOverlay, Polygon, LayersControl, Polyline, FullScreenControl, MarkerCluster, CircleMarker, Icon, AwesomeIcon
import matplotlib.pyplot as plt
from ipywidgets import HTML
import os

!jupyter nbextension enable --py widgetsnbextension --sys-prefix
!jupyter nbextension enable --py --sys-prefix ipyleaflet
```

```python
os.chdir("/Users/euan-soutter/Desktop/Python/Field/Azerbaijan/Data/") # Sets working directory
df = pd.read_csv("az_loc.csv")
df
```

```python
center = [41.2, 48.3]
zoom = 11
opacity = 0.5

# icon = AwesomeIcon(name=' fa-chevron-circle-down', icon_size = [5,5], icon_color='black')
icon = Icon(icon_url="https://img.icons8.com/material/24/000000/bookmark-ribbon--v1.png" ,icon_size=[5,5])

m = Map(basemap=basemaps.OpenTopoMap, center=center, zoom=zoom)
# m = Map(basemap=basemaps.OpenStreetMap.BlackAndWhite, center=center, zoom=zoom)

bochud = ImageOverlay(url='az_bochud.png', bounds=((41, 47), (41.92, 49)), opacity = opacity)
azmap = ImageOverlay(url='/Users/euan-soutter/Desktop/Python/Field/Azerbaijan/Data/az_map.png', bounds=((41, 48), (41.92, 49)), opacity = opacity)

dfn = df[['long', 'lat']]

locations = [tuple(r) for r in dfn.to_numpy()]

for i, k in locations:
    markers = Marker(location=(k, i), icon=icon)
    m.add_layer(markers)
    
# polygon = Polygon(locations= [[(40, 47),(41, 47),(41, 49),(40, 49)]], color="green", fill_color="green")

m.add_layer(bochud)
m.add_layer(azmap)
        
control = LayersControl(position='topright')
m.add_control(control)
m.add_control(FullScreenControl())

m
```
