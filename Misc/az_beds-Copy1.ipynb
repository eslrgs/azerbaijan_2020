```python
fig, axes = plt.subplots(ncols=3,nrows=2,figsize=(10,10))

colors = ['#48a2ba','lightseagreen','mediumseagreen','seagreen',"gold","sandybrown"]
cmap = mpl.colors.ListedColormap(colors)
cmap.set_under("grey")
cmap.set_over("indianred")
norm = mpl.colors.LogNorm(vmin=1, vmax=100)

edgecolors = 'none'
linewidth=0.2
    
def graph_bed(x, y, x_label, y_label, title, titleweight, titleloc, ax):
        ax=ax
        ax.scatter(x, y, c=x ,s=25,cmap=cmap,norm=norm,zorder=12,
        edgecolors=edgecolors,linewidths=linewidth)
        ax.plot(x, y,color='k',linestyle='dashed',linewidth=1)
        ax.set_ylabel(y_label,weight='bold')
        ax.set_xlim(left=0,right=650)
        ax.set_ylim(bottom=0)
        ax.set_xlabel(x_label,weight='bold')
        ax.set_title(title,weight=titleweight,pad=10,loc=titleloc)
        ax.grid(zorder=0,linestyle='dashed')
        
def graph_lith(x, y, x_label, y_label, title, ax):
        ax=ax
        ax.scatter(x, y, c=x ,s=25,cmap=cmap,zorder=12,
        edgecolors=edgecolors,linewidths=linewidth)
        ax.set_xticks(np.arange(1., 2.0001, step=1))
        ax.set_ylim(bottom=0)
        ax.set_xlim(0,3)
        ax.set_xticklabels(['Carbonate', 'Clastic'])
        ax.set_ylabel(y_label,weight='bold')
        ax.set_xlabel(x_label,weight='bold')
        ax.set_title(title,weight='bold',pad=10)
        ax.grid(zorder=0,linestyle='dashed')
        
def graph_amal(x, y, c, x_label, y_label, title, ax):
        ax=ax
        ax.scatter(x, y, c=c ,s=25,cmap=cmap,zorder=12,
        edgecolors=edgecolors,linewidths=linewidth)
        ax.set_ylabel(y_label,weight='bold')
        ax.set_ylim(bottom=0)
        ax.set_xticks(np.arange(0.5, 1.50001, step=0.5))
        ax.set_xlim(0.5,1.5)
        ax.set_xticklabels(['','Amalgamated'])
        ax.set_xlabel(x_label,weight='bold')
        ax.set_title(title,weight='bold',pad=10)
        ax.grid(zorder=0,linestyle='dashed')
        
def graph_gsize(x, y, x_label, y_label, title, ax):
        ax=ax
        ax.scatter(x, y, c=x ,s=25,cmap=cmap,zorder=12,
        edgecolors=edgecolors,linewidths=linewidth)
        ax.set_ylabel(y_label,weight='bold')
        ax.set_xlabel(x_label,weight='bold')
        ax.set_ylim(bottom=0)
        ax.set_xlim(2.5,8.5)
        ax.set_xticks(np.arange(3,8.0001,1))
        ax.set_xticklabels(['S', 'VF', 'F', 'M', 'C+', 'Cg'])
        ax.set_xlabel(x_label,weight='bold')
        ax.set_title(title,weight='bold',pad=10)
        ax.grid(zorder=0,linestyle='dashed')
        
graph_bed(df.beds_5C, df.Bed_Number,'Bed Thickness (cm)','Bed Number','N','bold','center',axes[0,0])
graph_bed(df.beds_4A, df.Bed_Number,'Bed Thickness (cm)','Bed Number','','normal','center', axes[0,1])
graph_bed(df.beds_3A, df.Bed_Number,'Bed Thickness (cm)','Bed Number','S','bold','center', axes[0,2])
# graph_lith(df.lith_5C, df.Bed_Number,'Lithology','Bed Number','', axes[1,0])
# graph_lith(df.lith_4A, df.Bed_Number,'Lithology','Bed Number','', axes[1,1])
# graph_lith(df.lith_3A, df.Bed_Number,'Lithology','Bed Number','', axes[1,2])
# graph_amal(df.amal_5C, df.Bed_Number,df.lith_5C,'Amalgamation','Bed Number','', axes[2,0])
# graph_amal(df.amal_4A, df.Bed_Number,df.lith_4A,'Amalgamation','Bed Number','', axes[2,1])
# graph_amal(df.amal_3A, df.Bed_Number,df.lith_3A,'Amalgamation','Bed Number','', axes[2,2])
graph_gsize(df.gsize_5C, df.Bed_Number,'Grain-Size','Bed Number','', axes[1,0])
graph_gsize(df.gsize_4A, df.Bed_Number,'Grain-Size','Bed Number','', axes[1,1])
graph_gsize(df.gsize_3A, df.Bed_Number,'Grain-Size','Bed Number','', axes[1,2])

plt.suptitle('Conacian-Maastrichtian Bed Statistics', y=1.03, x=0.531, fontsize=14, weight='bold')
plt.tight_layout()
plt.annotate('', xy=(0.250,0.937), xytext=(0.480,0.937), xycoords='figure fraction', 
             arrowprops=dict(arrowstyle='<->', facecolor='black'))
plt.annotate('', xy=(0.575,0.937), xytext=(0.815,0.937), xycoords='figure fraction', 
             arrowprops=dict(arrowstyle='<->', facecolor='black'))
plt.annotate('260 m', xy=(0.345,0.943), xycoords='figure fraction')
plt.annotate('490 m', xy=(0.675,0.943), xycoords='figure fraction')
plt.savefig('/Users/euan-soutter/Desktop/Python/Field_Data/Azerbaijan/Figs/az_beds_stats_2.png',
            dpi=400,bbox_inches='tight')
```

```python
fig, axes = plt.subplots(ncols=6,nrows=2,figsize=(15,8))

def r2(x, y):
    z = stats.pearsonr(x, y)[0] ** 2
    z = '%.2f' % z
    return z

def sns_reg(x, y, x_label, y_label, color, title, ax):
        ax=ax
        sns.regplot(x,y,color=color,ax=ax,scatter_kws={'s':5}) 
        ax.plot(x, y,color='k',linestyle='dashed',linewidth=0.5,zorder=0)
        ax.set_ylabel(y_label,weight='bold')
#       ax.set_xscale('log')
        if x.max() > 100:
            right = 100
        else:
            right = x.max()
        ax.set_xlim(left=0,right=right)
        ax.set_ylim(bottom=0,top=y.max())
        ax.set_xlabel(x_label,weight='bold')
        ax.set_title(title,weight='bold',pad=10,loc='center')
        ax.grid(zorder=0,linestyle='dashed')
        z = r2(x,y)
        ax.annotate(f'r\N{SUPERSCRIPT TWO} = {z}',(0.60,0.95),xycoords='axes fraction',fontsize=10,style='italic',color='r')

sns_reg(df5C_cal.beds_5C, df5C_cal.Bed_Number,'Bed Thickness (cm)','Bed Number','#48a2ba','Carbonate (5C)',axes[0,0])
sns_reg(df5C_sil.beds_5C, df5C_sil.Bed_Number,'Bed Thickness (cm)','Bed Number','sandybrown','Clastic (5C)',axes[0,1])

sns_reg(df4A_cal.beds_4A, df4A_cal.Bed_Number,'Bed Thickness (cm)','Bed Number',"#48a2ba",'Carbonate (4A)',axes[0,2])
sns_reg(df4A_sil.beds_4A, df4A_sil.Bed_Number,'Bed Thickness (cm)','Bed Number','sandybrown','Clastic (4A)',axes[0,3])

sns_reg(df3A_cal.beds_3A, df3A_cal.Bed_Number,'Bed Thickness (cm)','Bed Number',"#48a2ba",'Carbonate (3A)',axes[0,4])
sns_reg(df3A_sil.beds_3A, df3A_sil.Bed_Number,'Bed Thickness (cm)','Bed Number','sandybrown','Clastic (3A)',axes[0,5])

def sns_reg_g(x, y, x_label, y_label, color, title, ax1):
        ax=ax1
        sns.regplot(x,y,color=color,ax=ax,scatter_kws={'s':5}) 
        ax.plot(x, y, color='k',linestyle='dashed',linewidth=0.5,zorder=0)
        ax.set_ylabel(y_label,weight='bold')
        ax.set_xlim(2.5,8.5)
        ax.set_xticks(np.arange(3,8.0001,1))
        ax.set_xticklabels(['S', 'VF', 'F', 'M', 'C+', 'Cg'])
        ax.set_ylim(bottom=0,top=y.max())
        ax.set_xlabel(x_label,weight='bold')
        ax.set_title(title,weight='bold',pad=10,loc='center')
        ax.grid(zorder=0,linestyle='dashed')
        z = r2(x,y)
        ax.annotate(f'r\N{SUPERSCRIPT TWO} = {z}',(0.60,0.95),xycoords='axes fraction',fontsize=10,style='italic',color='r')
        
sns_reg_g(df5C_cal.gsize_5C, df5C_cal.Bed_Number,'Grain-Size','Bed Number',"#48a2ba",'Carbonate (5C)',axes[1,0])
sns_reg_g(df5C_sil.gsize_5C, df5C_sil.Bed_Number,'Grain-Size','Bed Number','sandybrown','Clastic (5C)',axes[1,1])

sns_reg_g(df4A_cal.gsize_4A, df4A_cal.Bed_Number,'Grain-Size','Bed Number',"#48a2ba",'Carbonate (4A)',axes[1,2])
sns_reg_g(df4A_sil.gsize_4A, df4A_sil.Bed_Number,'Grain-Size','Bed Number','sandybrown','Clastic (4A)',axes[1,3])

sns_reg_g(df3A_cal.gsize_3A, df3A_cal.Bed_Number,'Grain-Size','Bed Number',"#48a2ba",'Carbonate (3A)',axes[1,4])
sns_reg_g(df3A_sil.gsize_3A, df3A_sil.Bed_Number,'Grain-Size','Bed Number','sandybrown','Clastic (3A)',axes[1,5])

plt.suptitle('Carbonate/Clastic Thickness & Grain-Size Trends', y=1.03, x=0.518, fontsize=14, weight='bold')
plt.tight_layout()
plt.savefig('/Users/euan-soutter/Desktop/Python/Field_Data/Azerbaijan/Figs/az_cs_stats.png',
           dpi=400,bbox_inches='tight')
```

```python
from matplotlib.ticker import FuncFormatter
import matplotlib.ticker as mtick

fig,axs = plt.subplots(ncols=2,figsize=(11,5))

def plot_facies(log1,log2,log3,log1label,log2label,log3label,title,legtitle,ax):
        ax=ax
        ax.hist(log1, bins=np.arange(1.5,8.5,1),fill=False, 
         edgecolor = "indianred",linewidth=2,
        density=True,label=log1label,zorder=4)
        ax.hist(log2, bins=np.arange(1.5,8.5,1),fill=False,
         edgecolor = "dodgerblue",linewidth=2,
        density=True,label=log2label,zorder=3)
        ax.hist(log3, bins=np.arange(1.5,8.5,1),fill=False,
         edgecolor = "mediumseagreen",linewidth=2,linestyle=(0,(5,1)),
        density=True,label=log3label,zorder=2)
        ax.set_xlim(1.5,7.5)
        ax.set_ylim(0,1)
        ax.set_yticklabels(['0','20','40','60','80','100'])
        ax.set_xticklabels(['','HB', 'LDT', 'MDT', 'HDT', 'Db', 'Cg'])
        ax.set_xlabel('Facies', weight='bold')
        ax.set_ylabel('Proportion (%)',weight='bold')
        ax.grid(zorder=0,linestyle='dashed',axis='y')
        ax.legend(title=legtitle,fontsize=10,title_fontsize=12)
        ax.set_title(title)
        for axis in ['top','bottom','left','right']:
            ax.spines[axis].set_linewidth(1)
            ax.spines[axis].set_color("k")
            ax.spines[axis].set_zorder(10)
            
# plot_facies(df.facies_7D,df.facies_6A,df.facies_7D,'Off-axis','Axis','','Cenomanian - Turonian',ax=axs[0])
plot_facies(df.facies_7C,df.facies_7B,df.facies_7A,'Axis','Off-Axis','Overbank',
            'Cenomanian - Turonian','Channel',ax=axs[0])
plot_facies(df.facies_5A,df.facies_3A,df.facies_5C,'Axis','Off-Axis','Fringe', 
            'Conacian - Maastrictian','Lobe',ax=axs[1])

plt.suptitle('Upper Cretaceous Facies Statistics',weight='bold', x=0.515,y=1.0)
plt.savefig('/Users/euan-soutter/Desktop/Python/Field_Data/Azerbaijan/Figs/az_facies_stats.png',
           dpi=400,bbox_inches='tight')
```
